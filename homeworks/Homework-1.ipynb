```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_iris
from statistics import mean
```

```python
# Load iris data
iris = load_iris()

# Get feature data
dat = iris.data
```

```python
# Plot petal width by sepal width
plt.scatter(dat[:,1], dat[:,3])
plt.suptitle('Petal width by sepal width', fontsize=16)
plt.xlabel('sepal width (cm)', fontsize=14)
plt.ylabel('petal width (cm)', fontsize=14)
plt.grid(b=None, which='major', axis='both')
```

```python
# Convert data to pandas dataframe
dat1 = pd.DataFrame(dat)

# Add class labels
dat1 = dat1.assign(target = iris.target_names[iris.target])

# Rename column names
dat1.columns = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'Species']
```

```python
# Barplot of samples per species
sns.countplot(x="Species", data=dat1)
plt.suptitle('Samples per species', fontsize=16);
```

```python
# Histogram of sepal width
sns.distplot(dat1["sepal width (cm)"], kde=False)
plt.suptitle('Histogram of sepal width', fontsize=16)
plt.ylabel('count', fontsize=14);
```

```python
# Barplot of mean sepal length
sns.barplot(x="Species", y="sepal length (cm)", data=dat1, estimator=mean)
plt.suptitle('Mean sepal length by species', fontsize=16)
plt.ylabel('mean sepal length (cm)', fontsize=14);
```

```python
# Box plot of sepal width
sns.boxplot(x="Species", y="sepal width (cm)", data=dat1)
plt.suptitle('Sepal width by species', fontsize=16);
```

```python
# Violin plot of sepal width
sns.violinplot(x="Species", y="sepal width (cm)", data=dat1)
plt.suptitle('Sepal width by species', fontsize=16);
```

```python
# Alternat violin plot of sepal width
sns.violinplot(x="Species", y="sepal width (cm)", data=dat1, inner="quartile", scale="width")
plt.suptitle('Sepal width by species', fontsize=16);
```
