```python
import numpy as np
import pandas as pd
import tensorflow as tf
import time
import matplotlib.pyplot as plt
from datetime import datetime

# adjust width of notebook
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

### Load Data

```python
# load language sample
data = open('./alice.txt', 'r').read()

# unique characters and vocabulary length
chars = list(set(data))
chars.sort()
vocab_size = len(chars)

# map characters to numerics and vise versa
char_to_ix = {ch:i for i,ch in enumerate(chars)}
ix_to_char = {i:ch for i,ch in enumerate(chars)}
```

### Preprocess Data

```python
# number of recursive steps
seq_length = 25

# step sequence for plotting later
STEP_SEQUENCE = [i*50 for i in range(100)]

# training and target data
X_num, y_num = [], []

# translate data 
for i in range(0, len(data)-seq_length-1, 1):
    X_num.append([char_to_ix[ch] for ch in data[i:i+seq_length]])
    y_num.append([char_to_ix[ch] for ch in data[i+1:i+seq_length+1]])
    
# reshape data 
X_num = np.reshape(X_num, (len(X_num), seq_length))
y_num = np.reshape(y_num, (len(y_num), seq_length))

# tensorflow one hot encodeing
X_raw = tf.placeholder(tf.int32, [None, seq_length])
one_hot_X = tf.one_hot(X_raw, vocab_size)

# run session and one hot encode data
init = tf.global_variables_initializer()
with tf.Session() as sess:
    init.run()
    X_modified = one_hot_X.eval(feed_dict={X_raw:X_num})
```

## Function Definitions
###  Mini Batch Functions
These functions are for fetching sequential or random batches.

```python
def mini_batch_random(batch_size):
    i = np.random.randint(len(data)-seq_length-1, size=batch_size)
    X_batch = X_modified[i]
    y_batch = y_num[i]
    return X_batch, y_batch

def mini_batch_sequential(batch_size):
    i = np.random.randint(len(data)-seq_length-batch_size-1, size=1)[0]
    X_batch = X_modified[i:i+batch_size]
    y_batch = y_num[i:i+batch_size]
    return X_batch, y_batch
```

### Print Metrics
This function will print metrics nicely, given a training accuracy and training loss. The metrics are rounded to 5 significant figures, as the information loss is minimal.

```python
def print_metrics(step, training_acc, training_loss):
    # Round all metrics to 5 sigfigs, information loss is minimal.
    training_acc = round(training_acc, 5)
    training_loss = round(training_loss, 5)
    learn_rate = round(learning_schedule(step), 5)
    print('Step: {}\tTraining Accuracy: {}\tTraining Loss: {}\t\tLearning Rate: {}'.format(step, training_acc, training_loss, learn_rate))
```

### Plot Metrics
Make a plot of metrics given steps and points.

```python
def plot_metrics(title, ylabel, names, steps, *points):
    if len(names) != len(points):
        raise ValueError('Length of names ({}) is not equal to length of points array ({}).'.format(len(names), len(points)))
        
    _STEPS = steps
    colors = ["blue", "red", "green", "orange", "purple", "black", "gold", "brown"]   # I hope we don't need to plot more than 8 things
    
    plt.figure(figsize=(10,9))
    
    for i in range(len(names)):
        plt.plot(_STEPS, points[i], colors[i], label=names[i])
    
    plt.ylabel(ylabel)
    plt.xlabel("Steps")
    plt.title(title)
    
    plt.legend()
    plt.show()
```

### Define Learning Schedule
Define a learning schedule, parameters can be tweaked for a steeper or more gradual learning rate.

```python
def learning_schedule(step=0, start=0.011, stop=0.005, decay=1000):
    return start/(1 + step/decay) + stop

# compute image of learning schedule
image = []
for i in np.arange(10000):
    image.append(learning_schedule(step=i))

# plot learning schedule
plt.figure(figsize=(10, 5))
plt.title('Learning schedule (default)', fontsize=18)
plt.xlabel('step', fontsize=16)
plt.ylabel('learning rate', fontsize=16)
plt.plot(image);
```

## Matching Karpathy's RNN
In the following cells, we will be attempting to mimic Karpathy's handwritten recursive neural net. We keep the hidden size at 100 neurons, and use tanh similar to Karpathy. We then train and sample the output from the model using both sequential batches and random batches. Sequential in this case is similar to Karpathy, where chunks of contiguous characters are sequentially fed to the model. Random takes chunks of contiguous characters from random spots of the text. With some testing, we decided not having an OutputProjectionLayer gave the model better performance, so we decided to discontinue using it.

For all our models we decided to keep `n_epochs`, `iterations`, and `batch_size` constant at 5, 1000, and 1000 respectively.

### Construction Phase


```python
# training hyperparamters
n_epochs = 5
iterations = 1000
batch_size = 1000

# network hyperparamters
n_inputs = vocab_size
n_neurons = 100
n_outputs = 1

# define network structure
tf.reset_default_graph()

# for sample function
keep_rate = tf.placeholder(tf.float64)
X_raw = tf.placeholder(tf.int32, [None, seq_length])
one_hot_X = tf.one_hot(X_raw, vocab_size)

# placeholders for data and targets
X = tf.placeholder(tf.float64, [None, seq_length, n_inputs])
y = tf.placeholder(tf.int32, [None, seq_length])

# basic recursive cell
cell = tf.nn.rnn_cell.BasicRNNCell(num_units=n_neurons, activation=tf.nn.tanh)

# save outputs and states
outputs, states = tf.nn.dynamic_rnn(cell, X, dtype=tf.float64)

# define loss function
logit = tf.contrib.layers.fully_connected(outputs, vocab_size, activation_fn=None)
loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logit))

# optimize loss function
learning_rate = tf.placeholder(tf.float64)
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)

# clip and apply gradiants
grads = optimizer.compute_gradients(loss)
capped_grads = [(tf.clip_by_value(grad, -5, 5), var) for grad, var in grads]
training_op = optimizer.apply_gradients(capped_grads)

# compute prediction and accuracy
prediction = tf.nn.softmax(logit)
correct = tf.equal(tf.cast(y, tf.int64), tf.argmax(prediction, 2))
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

# allow gpu growth
config = tf.ConfigProto()
config.gpu_options.allow_growth = True

# intialize variables
init = tf.global_variables_initializer()
```

### Training with Random Mini Batches

```python
# metrics for plotting after
KARPATHY_RANDOM = []

# saver for model
saver = tf.train.Saver()

# coefficient for smoothing loss and accuracy
smoothing_coef = 0.5

# run session
with tf.Session() as sess:
    init.run()

    # smooth training loss and accuracy at step 0
    smooth_train_loss = -np.log(1.0 / vocab_size) * seq_length
    smooth_train_accuracy = 0
    step = 0

    # train for n_epochs times iterations steps
    for epoch in range(n_epochs):
        for iteration in range(iterations):
            step += 1
                
            # fetch mini batch and run train operation
            X_batch, y_batch = mini_batch_random(batch_size)
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch, learning_rate: learning_schedule(step)}) 
            
            # compute smooth training loss and accuracy
            if iteration % 50 == 0:
                smooth_train_loss = smooth_train_loss * smoothing_coef + loss.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)
                smooth_train_accuracy = smooth_train_accuracy * smoothing_coef + accuracy.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)
        
                # add point to metrics
                KARPATHY_RANDOM.append(smooth_train_accuracy)

        # periodically display accuracy, smooth loss, and text sample. Metrics rounded to 5 sigfigs.
        print_metrics(step, smooth_train_accuracy, smooth_train_loss)
        sample()
              
    save_path = saver.save(sess, "./random_model")
```

### Training with Sequential Mini Batches

```python
# metrics for plotting after
KARPATHY_SEQUENTIAL = []

# saver for model
saver = tf.train.Saver()

# coefficient for smoothing loss and accuracy
smoothing_coef = 0.5

# run session
with tf.Session() as sess:
    init.run()
    
   # smooth training loss and accuracy at step 0
    smooth_train_loss = -np.log(1.0 / vocab_size) * seq_length
    smooth_train_accuracy = 0
    step = 0

    # train for n_epochs times iterations steps
    for epoch in range(n_epochs):
        for iteration in range(iterations):
            step += 1
                
            # fetch mini batch and run train operation
            X_batch, y_batch = mini_batch_sequential(batch_size)
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch, learning_rate: learning_schedule(step)}) 
            
            # compute smooth training loss and accuracy
            if iteration % 50 == 0:
                smooth_train_loss = smooth_train_loss * smoothing_coef + loss.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)
                smooth_train_accuracy = smooth_train_accuracy * smoothing_coef + accuracy.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)
        
                # add point to metrics
                KARPATHY_SEQUENTIAL.append(smooth_train_accuracy)
        
        # periodically display accuracy, smooth loss, and text sample. Metrics rounded to 5 sigfigs.
        print_metrics(step, smooth_train_accuracy, smooth_train_loss)
        sample()
              
    # save model
    save_path = saver.save(sess, "./sequential_model")
```

## Gated Recurrent Unit Model
Next, we observed changing the type of cell to a gated recurrent unit. We are utilizing the GPU enabled version of this cell, and we experienced immediately better results than using a basic cell. To more accurately compare these other cells to Karpathy's code, we decided to stick to 100 neurons and only change the type of cell.

### Construction Phase

```python
# training hyperparamters
n_epochs = 5
iterations = 1000
batch_size = 1000

# network hyperparamters
n_inputs = vocab_size
n_neurons = 100
n_outputs = 1

# define network structure
tf.reset_default_graph()
he_init = tf.contrib.layers.variance_scaling_initializer(mode="FAN_AVG")

# for sample function
keep_rate = tf.placeholder(tf.float64)
X_raw = tf.placeholder(tf.int32, [None, seq_length])
one_hot_X = tf.one_hot(X_raw, vocab_size)

# placeholders for data and targets
X = tf.placeholder(tf.float32, [None, seq_length, n_inputs])
y = tf.placeholder(tf.int32, [None, seq_length])

# GRU cell within projection wrapper
cell = tf.contrib.cudnn_rnn.CudnnCompatibleGRUCell(num_units=n_neurons, kernel_initializer=he_init)

# save outputs and states
outputs, states = tf.nn.dynamic_rnn(cell=cell, inputs=X, dtype=tf.float32)
        
# define loss function
logit = tf.contrib.layers.fully_connected(outputs, vocab_size, activation_fn=None)
logit = tf.cast(logit, tf.float64)
loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logit))

# optimize loss function
learning_rate = tf.placeholder(tf.float64)
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)

# clip and apply gradiants
grads = optimizer.compute_gradients(loss)
capped_grads = [(tf.clip_by_value(grad, -5, 5), var) for grad, var in grads]
training_op = optimizer.apply_gradients(capped_grads)

# compute prediction and accuracy
prediction = tf.nn.softmax(logit)
correct = tf.equal(tf.cast(y, tf.int64), tf.argmax(prediction, 2))
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

# allow gpu growth
config = tf.ConfigProto()
config.gpu_options.allow_growth = True

# intialize variables
init = tf.global_variables_initializer()
```

### Training GRU Model

```python
# metrics for plotting later
GRU_RANDOM = []

# saver for model
saver = tf.train.Saver()

# coefficient for smoothing loss and accuracy
smoothing_coef = 0.5

# run session
with tf.Session() as sess:
    init.run()

    # smooth training loss and accuracy at step 0
    smooth_train_loss = -np.log(1.0 / vocab_size) * seq_length
    smooth_train_accuracy = 0
    step = 0

    # train for n_epochs times iterations steps
    for epoch in range(n_epochs):
        for iteration in range(iterations):
            step += 1
                
            # fetch mini batch and run train operation
            X_batch, y_batch = mini_batch_random(batch_size)
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch, learning_rate: learning_schedule(step)}) 
            
            # compute smooth training loss and accuracy
            if iteration % 50 == 0:
                smooth_train_loss = smooth_train_loss * smoothing_coef + loss.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)
                smooth_train_accuracy = smooth_train_accuracy * smoothing_coef + accuracy.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)

                # add point to metrics
                GRU_RANDOM.append(smooth_train_accuracy)
        
        # periodically display accuracy, smooth loss, and text sample. Metrics rounded to 5 sigfigs.
        print_metrics(step, smooth_train_accuracy, smooth_train_loss)
        sample()
              
    # save model
    save_path = saver.save(sess, "./gru_model")
```

## Grid Searching for Final Model
Previously, we observed that LSTM seemed to be better than GRU cells. However, to find the optimal parameters for each, we decided to set up a manual grid search that was modular, and allowed us to test any combination of parameters we wanted to. Below, we define the class to hold parameter options, and set up the manual grid search.

### Construction
While we are performing this manual grid search, we have a few parameters that will remain constant.
- Number of epochs: 5
- Number of iterations: 1000
- Batch size: 1000
- Smoothing coefficient: 0.5

The parameters we will be testing in this grid search will be `MODEL_TYPE`, `NUM_NEURONS` (afterwards), `NUM_LAYERS`, `LEARNING_RATE_DECAY`.
Accuracy metrics from each model will be stored every 50 iterations, resulting in 100 total points for each model. Output will not be generated during this grid search. Only the training accuracy and loss will be recorded for each model for comparison after the grid search.

```python
class ParamSet:
    def __init__(self, model_name='none', model_is_gru=False, num_neurons=200, num_layers=3, decay_rate=5000):
        self.name = model_name
        self.model_is_GRU = model_is_gru
        self.num_neurons = num_neurons
        self.num_layers = num_layers
        self.decay_rate = decay_rate
        self.acc_metrics = []
        self.loss_metrics = []
        
    def print_params(self):
        print('Model name: {}\nModel is GRU?: {}\nNumber of neurons: {}\nNumber of layers: {}\nDecay rate: {}\n'.format(
            self.name, self.model_is_GRU, self.num_neurons, self.num_layers, self.decay_rate))
        
    def print_metrics(self):
        if len(self.acc_metrics) == 0 or len(self.loss_metrics) == 0:
            print("[ParamSet]: Cannot print metrics with no accuracy or loss metrics.")
            return
        
        acc = round(self.acc_metrics[-1], 4)
        loss = round(self.loss_metrics[-1], 4)
        name = self.name
        highest_step = np.argmax(self.acc_metrics)

        print("-- {} --".format(name))
        print("Step with highest accuracy:\t{}\nAcc @ Step {}:\t\t\t{}".format(
            highest_step, highest_step, round(self.acc_metrics[highest_step], 4)))
        print("Ending accuracy:\t\t{}".format(acc))
        print("Ending loss:\t\t\t{}\n".format(loss))
    
    def get_last_accuracy(self):
        return self.acc_metrics[-1]
    
    def get_last_loss(self):
        return self.loss_metrics[-1]
        
    def add_accuracy_point(self, point):
        self.acc_metrics.append(point)
    
    def add_loss_point(self, point):
        self.loss_metrics.append(point)
```

### Grid Search Function

```python
def manual_grid_search(param_set_list):
    # constant training hyperparameters
    n_epochs = 5
    iterations = 1000
    batch_size = 1000

    # constant network hyperparamters
    n_inputs = vocab_size
    n_outputs = 1

    print("[ManualGS]: --- Starting manual grid search! ---")
    for param_set in param_set_list:
        start = time.time()

        # get parameters from object
        MODEL_IS_GRU = param_set.model_is_GRU
        NUM_NEURONS = param_set.num_neurons
        NUM_LAYERS = param_set.num_layers
        DECAY_RATE = param_set.decay_rate

        # define network structure
        tf.reset_default_graph()
        he_init = tf.contrib.layers.variance_scaling_initializer(mode="FAN_AVG")

        # for sample function
        keep_rate = tf.placeholder(tf.float64)
        X_raw = tf.placeholder(tf.int32, [None, seq_length])
        one_hot_X = tf.one_hot(X_raw, vocab_size)

        # placeholders for data and targets
        X = tf.placeholder(tf.float32, [None, seq_length, n_inputs])
        y = tf.placeholder(tf.int32, [None, seq_length])

        # stacked cells construction
        num_layers = NUM_LAYERS
        if MODEL_IS_GRU:
            stacked_rnn = tf.contrib.rnn.MultiRNNCell(
                [tf.contrib.cudnn_rnn.CudnnCompatibleGRUCell(num_units=NUM_NEURONS) for _ in range(num_layers)]
            )
        else:
            stacked_rnn = tf.contrib.rnn.MultiRNNCell(
                [tf.contrib.cudnn_rnn.CudnnCompatibleLSTMCell(num_units=NUM_NEURONS) for _ in range(num_layers)]
            )

        # save outputs and states
        final_outputs, final_state = tf.nn.dynamic_rnn(cell=stacked_rnn, inputs=X, dtype=tf.float32)

        # define loss function
        logit = tf.contrib.layers.fully_connected(final_outputs, vocab_size, activation_fn=None)
        logit = tf.cast(logit, tf.float64)
        loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logit))

        # optimize loss function
        learning_rate = tf.placeholder(tf.float64)
        optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)

        # clip and apply gradiants
        grads = optimizer.compute_gradients(loss)
        capped_grads = [(tf.clip_by_value(grad, -5, 5), var) for grad, var in grads]
        training_op = optimizer.apply_gradients(capped_grads)

        # compute prediction and accuracy
        prediction = tf.nn.softmax(logit)
        correct = tf.equal(tf.cast(y, tf.int64), tf.argmax(prediction, 2))
        accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

        # allow gpu growth
        config = tf.ConfigProto()
        config.gpu_options.allow_growth = True

        # intialize variables
        init = tf.global_variables_initializer()

        # coefficient for smoothing loss and accuracy
        smoothing_coef = 0.5

        # run session
        with tf.Session() as sess:
            init.run()

            # smooth training loss and accuracy at step 0
            smooth_train_loss = -np.log(1.0 / vocab_size) * seq_length
            smooth_train_accuracy = 0
            step = 0

            # train for n_epochs times iterations steps
            for epoch in range(n_epochs):
                for iteration in range(iterations):
                    step += 1

                    # fetch mini batch and run train operation
                    X_batch, y_batch = mini_batch_random(batch_size)
                    sess.run(training_op, feed_dict={X: X_batch, y: y_batch, learning_rate: learning_schedule(step, decay=DECAY_RATE)}) 

                    # compute smooth training loss and accuracy
                    if iteration % 50 == 0:
                        smooth_train_loss = smooth_train_loss * smoothing_coef + loss.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)
                        smooth_train_accuracy = smooth_train_accuracy * smoothing_coef + accuracy.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)

                        # add point to metrics every 50 iterations
                        param_set.add_accuracy_point(smooth_train_accuracy)
                        param_set.add_loss_point(smooth_train_loss)

        end = time.time()
        print('[ManualGS]: Training complete for {}!\nElapsed time: {} seconds.\n---\n'.format(param_set.name, round(end-start, 2)))

    print("[ManualGS]: --- Grid Search Complete! ---")
```

## Building the Final Model
Using what we discovered from the grid searches, we will build our real final model and test it out! Since this is our final model, we will run it for 10,000 steps instead of 5,000, and then sample the output a few times towards the end.

```python
# set params for graph
FM_NUM_NEURONS = 200
FM_NUM_LAYERS = 4
FM_DECAY_RATE = 100

# training hyperparamters
n_epochs = 10
iterations = 100
batch_size = 1000

# network hyperparamters
n_inputs = vocab_size
n_outputs = 1

# define network structure
tf.reset_default_graph()
he_init = tf.contrib.layers.variance_scaling_initializer(mode="FAN_AVG")

# for sample function
keep_rate = tf.placeholder(tf.float64)
X_raw = tf.placeholder(tf.int32, [None, seq_length])
one_hot_X = tf.one_hot(X_raw, vocab_size)

# placeholders for data and targets
X = tf.placeholder(tf.float32, [None, seq_length, n_inputs])
y = tf.placeholder(tf.int32, [None, seq_length])

# stacked cells construction, using LSTM
num_layers = FM_NUM_LAYERS
stacked_rnn = tf.contrib.rnn.MultiRNNCell(
    [tf.contrib.cudnn_rnn.CudnnCompatibleLSTMCell(num_units=FM_NUM_NEURONS) for _ in range(num_layers)]
)

# save outputs and states
final_outputs, final_state = tf.nn.dynamic_rnn(cell=stacked_rnn, inputs=X, dtype=tf.float32)
        
# define loss function
logit = tf.contrib.layers.fully_connected(final_outputs, vocab_size, activation_fn=None)
logit = tf.cast(logit, tf.float64)
loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logit))

# optimize loss function
learning_rate = tf.placeholder(tf.float64)
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)

# clip and apply gradiants
grads = optimizer.compute_gradients(loss)
capped_grads = [(tf.clip_by_value(grad, -5, 5), var) for grad, var in grads]
training_op = optimizer.apply_gradients(capped_grads)

# compute prediction and accuracy
prediction = tf.nn.softmax(logit)
correct = tf.equal(tf.cast(y, tf.int64), tf.argmax(prediction, 2))
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

# allow gpu growth
config = tf.ConfigProto()
config.gpu_options.allow_growth = True

# intialize variables
init = tf.global_variables_initializer()
```

### Training Final Model

```python
# metrics for plotting later
FINAL_MODEL_ACC, FINAL_MODEL_LOSS = [], []

# saver for model
saver = tf.train.Saver()

# coefficient for smoothing loss and accuracy
smoothing_coef = 0.5

# run session
with tf.Session() as sess:
    init.run()

    # smooth training loss and accuracy at step 0
    smooth_train_loss = -np.log(1.0 / vocab_size) * seq_length
    smooth_train_accuracy = 0
    step = 0

    # train for n_epochs times iterations steps
    for epoch in range(n_epochs):
        for iteration in range(iterations):
            step += 1
                
            # fetch mini batch and run train operation
            X_batch, y_batch = mini_batch_random(batch_size)
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch, learning_rate: learning_schedule(step, decay=FM_DECAY_RATE)}) 
            
        # smooth metrics
        smooth_train_loss = smooth_train_loss * smoothing_coef + loss.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)
        smooth_train_accuracy = smooth_train_accuracy * smoothing_coef + accuracy.eval(feed_dict={X: X_batch, y: y_batch}) * (1-smoothing_coef)
                
        # add points to metrics
        FINAL_MODEL_ACC.append(smooth_train_accuracy)
        FINAL_MODEL_LOSS.append(smooth_train_loss)
        
        # periodically display accuracy, smooth loss, and text sample. Metrics rounded to 5 sigfigs.
        print_metrics(step, smooth_train_accuracy, smooth_train_loss)
        sample()
              
    # save model
    saver.save(sess, "./final_model.ckpt")
```

We found that translating a handwritten recursive neural net into TensorFlow code was not as simple as it first seemed, and that there were many technical challenges to solve along the way. Our initial model that mimicked the handwritten model did not perform very well, but as we tweaked the model parameters, we experienced significantly increased accuracy and results. We found that LSTM cells seem to work very well. We also did some research into Google recommended structures for these kinds of neural nets while we were working, and found one paper which claimed that 4 layers seemed to be better for LSTM cells. From our testing conducted above with this problem, it seems that the abstract was indeed correct that 4 layer LSTMs work very well! However, the difference between that and a 3 layer LSTM was marginal.
