### Cumulative Explained Variance Ratio

```python
pca = PCA()
pca.fit(dat)
cumsum = np.cumsum(pca.explained_variance_ratio_)
plt.plot(cumsum);
```

### Visualizing Data with Two Principal Components

```python
pca = PCA(n_components=2)
dat_PCA = pd.concat([pd.DataFrame(pca.fit_transform(dat)), dat["default.payment.next.month"]], axis=1)
sns.scatterplot(x=0, y=1, hue="default.payment.next.month", data=dat_PCA);
```
